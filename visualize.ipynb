```python
import numpy as np
import matplotlib.pyplot as plt

# Define the periodic kernel function
def periodic_kernel(x_a, x_b, sigma2=1.0, l=1.0, p=1.0):
    return sigma2 * np.exp(-2 * (np.sin(np.pi * np.abs(x_a - x_b) / p) ** 2) / (l ** 2))

# Create a grid of points
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)

# Compute the kernel function for each pair of points
Z = periodic_kernel(X, Y)

# Plot the 3D visualization
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax.set_xlabel('x_a')
ax.set_ylabel('x_b')
ax.set_zlabel('k(x_a, x_b)')
ax.set_title('Periodic Kernel Visualization')
fig.colorbar(surf, shrink=0.5, aspect=10)
# plt.show()
plt.savefig('periodic_kernel.png')
```

```python
# Define the modified kernel function
def modified_kernel(x_a, x_b, sigma2=1.0, l_p=1.0, l_eq=1.0, p=1.0):
    periodic_term = np.exp(-2 * (np.sin(np.pi * np.abs(x_a - x_b) / p) ** 2) / (l_p ** 2))
    rbf_term = np.exp(-np.linalg.norm(x_a - x_b)**2 / (2 * l_eq ** 2))
    return sigma2 * periodic_term * rbf_term

# Create a grid of points
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)

# Compute the modified kernel function for each pair of points
Z = np.zeros_like(X)
for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        Z[i, j] = modified_kernel(X[i, j], Y[i, j])

# Plot the 3D visualization
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax.set_xlabel('x_a')
ax.set_ylabel('x_b')
ax.set_zlabel('k(x_a, x_b)')
ax.set_title('Modified Kernel Visualization')
fig.colorbar(surf, shrink=0.5, aspect=10)
# plt.show()
plt.savefig('modified_kernel.png')
```

```python
# Define the Rational Quadratic kernel function
def rational_quadratic_kernel(x_a, x_b, sigma2=1.0, alpha=1.0, l=1.0):
    return sigma2 * (1 + np.linalg.norm(x_a - x_b) ** 2 / (2 * alpha * l ** 2)) ** (-alpha)

# Create a grid of points
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)

# Compute the Rational Quadratic kernel function for each pair of points
Z = np.zeros_like(X)
for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        Z[i, j] = rational_quadratic_kernel(X[i, j], Y[i, j])

# Plot the 3D visualization
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax.set_xlabel('x_a')
ax.set_ylabel('x_b')
ax.set_zlabel('k(x_a, x_b)')
ax.set_title('Rational Quadratic Kernel Visualization')
fig.colorbar(surf, shrink=0.5, aspect=10)
# plt.show()
plt.savefig('rational_quadratic_kernel.png')
```
